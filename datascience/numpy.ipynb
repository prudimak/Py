```python
import matplotlib.pyplot as plt
```

```python
x=np.array([50,40,160,110])
my_lable=(["jeep","bmw","honda","toyota"])
my_explode=[0.3,0,0,0]

plt.pie(x, labels=my_lable, explode=my_explode, shadow=True)
plt.legend(title="cars")
plt.show()
```
